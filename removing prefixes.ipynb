```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('dataset.csv', encoding='cp1252',names=['sites'])
data.head()
```

```python
data.to_csv('new_dataset.csv',columns=['sites'],index=False,)
new_data = pd.read_csv('new_dataset.csv')
new_data
```

```python
new_data = new_data.sites.apply(lambda x: x.rstrip('http//').replace('www.', ' ').replace('http://', ' ').replace('\nhttps', ' ').replace('://', ' ').replace('\n', '').replace('\xa0', '').replace('/', ''))
new_gotten_data = pd.DataFrame(new_data,columns=['sites'])
new_gotten_data
```

```python
new_gotten_data.to_csv('new_gotten_data.csv',columns=['sites'],index=False)
```
